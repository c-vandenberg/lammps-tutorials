```python
import numpy as numpy
from numpy import ndarray
from matplotlib.legend import Legend
from matplotlib import pyplot as pyplot
```

# Extract stress & strain data for unbreakable & breakable bonds CNTs. Convert stress to GPa

```python
cnt_unbreakable_bonds_path: str = '../cnt-unbreakable-bonds/data/raw/stress-strain/'

# File contains columns of data, the .T transpose operation switches them to rows, making it easier to
# unpack into separate variables. Ignore time with `_` placeholder
# Extract CNT unbreakable bonds strain
_, cnt_unbreakable_bonds_strain = numpy.loadtxt(
    cnt_unbreakable_bonds_path + 'output_cnt_strain.dat'
).T

# Extract CNT unbreakable bonds stress
_, cnt_unbreakable_bonds_stress = numpy.loadtxt(
    cnt_unbreakable_bonds_path + 'output_cnt_stress.dat'
).T

# Convert stress from Pa to GPa
cnt_unbreakable_bonds_stress /= 1e9

cnt_breakable_bonds_path: str = '../cnt-breakable-bonds/data/raw/stress-strain/'

# Extract CNT breakable bonds strain
_, cnt_breakable_bonds_strain = numpy.loadtxt(
    cnt_breakable_bonds_path + 'output_cnt_strain.dat'
).T

# Extract CNT breakable bonds stress
_, cnt_breakable_bonds_stress = numpy.loadtxt(
    cnt_breakable_bonds_path + 'output_cnt_stress.dat'
).T

# Convert strain from Pa to GPa
cnt_breakable_bonds_stress /= 1e9
```

# Create Matplotlib graph to plot stress v strain for unbreakable & breakable bonds CNTs

```python
# Set up the line graph
cnt_stress_v_strain_line_graph, cnt_stress_v_strain_line_graph_axes = pyplot.subplots(figsize=(18, 10))
cnt_stress_v_strain_line_graph.patch.set_facecolor('black')
cnt_stress_v_strain_line_graph_axes.set_facecolor('black')

# Stylise ticks and spines
cnt_stress_v_strain_line_graph_axes.tick_params(colors='white', which='both', labelsize=20)
for spine in cnt_stress_v_strain_line_graph_axes.spines.values():
    spine.set_edgecolor('white')

# Plot the data
cnt_stress_v_strain_line_graph_axes.plot(cnt_unbreakable_bonds_strain, cnt_unbreakable_bonds_stress, color='cyan', 
                                linewidth=3.5, label='CNT - Unbreakable Bonds')
cnt_stress_v_strain_line_graph_axes.plot(cnt_breakable_bonds_strain, cnt_breakable_bonds_stress, color='orange', 
                                linewidth=3.5, label='CNT - Breakable Bonds')

# Set axes labels, limits and graph title
cnt_stress_v_strain_line_graph_axes.set_xlabel(r'$Strain$ (%)', color='white', fontsize=20)
cnt_stress_v_strain_line_graph_axes.set_ylabel(r'$Stress$ (GPa)', color='white', fontsize=20)
cnt_stress_v_strain_line_graph_axes.set_xlim(-1, 35)
cnt_stress_v_strain_line_graph_axes.set_ylim(-50, 650)
cnt_stress_v_strain_line_graph_axes.set_title(r'$\bf{Unbreakable\ &\ Breakable Bonds\ Carbon\ Nanotube\ (CNT)\ Stress\ vs\ Strain}$', 
                                     fontsize=20, color='white')

# Add legend
legend: Legend = cnt_stress_v_strain_line_graph_axes.legend(loc='upper left', frameon=False, fontsize=17.5)
for text in legend.get_texts():
    text.set_color('white')

# Add figure title
cnt_stress_v_strain_line_graph.text(0.5, 0.005,
                r'$\bf{Fig\ 1}$ Stress-strain curves for unbreakable & breakable CNTs',
                ha='center', va='center', color='white', fontsize=20)

# Save & display the plot
pyplot.savefig('../reports/figures/graphs/stress_vs_strain.png', bbox_inches='tight')
pyplot.show()
```
