```python
import numpy as numpy
from numpy import ndarray
from matplotlib.legend import Legend
from matplotlib import pyplot as pyplot
```

# Extract time and PEG length data from `ave_PEG_end_to_end_distance.dat` file and convert to ps & nm respectively

```python
# Unpack time and PEG length data
time, peg_length = numpy.loadtxt(
    '../data/raw/length-vs-time/ave_PEG_end_to_end_distance.dat'
).T

# Convert time from femtoseconds to picoseconds
time /= 1000

# Convert PEG length data from Ã… (10e-10) to nm (10e-9)
peg_length /= 10
```

# Create Matplotlib graph to plot PEG end-to-end length vs time

```python
# Set up the line graph
peg_length_vs_time_line_graph, peg_length_vs_time_line_graph_axes = pyplot.subplots(figsize=(18, 6))
peg_length_vs_time_line_graph.patch.set_facecolor('black')
peg_length_vs_time_line_graph_axes.set_facecolor('black')

# Stylise ticks and spines
peg_length_vs_time_line_graph_axes.tick_params(colors='white', which='both', labelsize=20)
for spine in peg_length_vs_time_line_graph_axes.spines.values():
    spine.set_edgecolor('white')

# Plot the data
peg_length_vs_time_line_graph_axes.plot(time, peg_length, color='cyan', linewidth=3.5)

# Add dashed line at x=31.5
peg_length_vs_time_line_graph_axes.axvline(31.5, color='white', linestyle='--', linewidth=1.5)

# Set axes labels, limits and graph title
peg_length_vs_time_line_graph_axes.set_xlabel('$t$ (ps)', fontsize=20, color='white')
peg_length_vs_time_line_graph_axes.set_ylabel(r'$PEG\ Length$ (nm)', fontsize=20, color='white')
peg_length_vs_time_line_graph_axes.set_xlim(0, 60)
peg_length_vs_time_line_graph_axes.set_ylim(0, 6)
peg_length_vs_time_line_graph_axes.set_title(r'$\bf{PEG\ Length\ (End-to-End)\ vs\ Time}$', 
                                             fontsize=20, color='white')

# Add figure title
peg_length_vs_time_line_graph.text(0.5, -0.075,
                r'$\bf{Fig\ 1}$ Evolution of end-to-end PEG length as a function of time. The force is applied at '
                r'~ $t$ = 31.5',
                ha='center', va='center', color='white', fontsize=20)

# Save & display the plot
pyplot.savefig('../reports/figures/graphs/peg-length-vs-time.png', bbox_inches='tight')
pyplot.show()
```
