```python
import numpy as numpy
from numpy import ndarray
from matplotlib.legend import Legend
from matplotlib import pyplot as pyplot
```

# Extract pure H2O time, temperature, volume & density data

```python
 # File contains columns of data, the .T transpose operation switches them to rows, making it easier to
# unpack into separate variables
# Extract H2O temperature data
time, h2o_temperature = numpy.loadtxt(
    '../data/raw/temperature-vs-time/ave_H2O_temperature.dat'
).T

# Subtracts the first element from every element in the time array, ensuring the first element is 0/the time starts
# at zero
time -= time[0]

# Convert time from femtoseconds to picoseconds
time /= 1000

# Extract H2O volume data
_, h2o_volume = numpy.loadtxt(
    '../data/raw/volume-vs-time/ave_H2O_volume.dat'
).T

# Convert H2O volume from cubic Å to cubic nm
h2o_volume /= 1000

# Extract time and H2O density data
_, h2o_density = numpy.loadtxt(
    '../data/raw/density-vs-time/ave_H2O_density.dat'
).T

# Convert density from molecule/Å-3 to molecule/nm-3
h2o_density *= 1000
```

# Create Matplotlib graph to plot pure H2O temperature vs time

```python
# Set up the line graph
temperature_vs_time_line_graph, temperature_vs_time_line_graph_axes = pyplot.subplots(figsize=(18, 10))
temperature_vs_time_line_graph.patch.set_facecolor('black')
temperature_vs_time_line_graph_axes.set_facecolor('black')

# Stylise ticks and spines
temperature_vs_time_line_graph_axes.tick_params(colors='white', which='both', labelsize=20)
for spine in temperature_vs_time_line_graph_axes.spines.values():
    spine.set_edgecolor('white')

# Plot the data
temperature_vs_time_line_graph_axes.plot(time, h2o_temperature, color='cyan', 
                                linewidth=3.5)

# Set axes labels, limits and graph title
temperature_vs_time_line_graph_axes.set_xlabel(r'$t$ (ps)', color='white', fontsize=20)
temperature_vs_time_line_graph_axes.set_ylabel(r'$Temperature$ (K)', color='white', fontsize=20)
temperature_vs_time_line_graph_axes.set_xlim(0, 20)
temperature_vs_time_line_graph_axes.set_ylim(25, 325)
temperature_vs_time_line_graph_axes.set_title(
    r'$\bf{Pure\ H_{2}O\ Temperature\ vs\ Time}$', 
    fontsize=25, color='white'
)

# Add figure title
temperature_vs_time_line_graph.text(0.5, 0.005,
                r'$\bf{Fig\ 1}$ Evolution of pure H$_{2}$O temperature as a function of time',
                ha='center', va='center', color='white', fontsize=20)

# Save & display the plot
pyplot.savefig('../reports/figures/graphs/pure_H2O_temp_vs_time.png', bbox_inches='tight')
pyplot.show()
```

# Create Matplotlib graph to plot pure H2O volume vs time

```python
# Set up the line graph
volume_vs_time_line_graph, volume_vs_time_line_graph_axes = pyplot.subplots(figsize=(18, 10))
volume_vs_time_line_graph.patch.set_facecolor('black')
volume_vs_time_line_graph_axes.set_facecolor('black')

# Stylise ticks and spines
volume_vs_time_line_graph_axes.tick_params(colors='white', which='both', labelsize=20)
for spine in volume_vs_time_line_graph_axes.spines.values():
    spine.set_edgecolor('white')

# Plot the data
volume_vs_time_line_graph_axes.plot(time, h2o_volume, color='cyan', 
                                linewidth=3.5)

# Set axes labels, limits and graph title
volume_vs_time_line_graph_axes.set_xlabel(r'$t$ (ps)', color='white', fontsize=20)
volume_vs_time_line_graph_axes.set_ylabel(r'$Volume$ (nm$^{3}$)', color='white', fontsize=20)
volume_vs_time_line_graph_axes.set_xlim(0, 20)
volume_vs_time_line_graph_axes.set_ylim(30, 90)
volume_vs_time_line_graph_axes.set_title(
    r'$\bf{Pure\ H_{2}O\ Volume\ vs\ Time}$', 
    fontsize=25, color='white'
)

# Add figure title
volume_vs_time_line_graph.text(0.5, 0.005,
                r'$\bf{Fig\ 2}$ Evolution of pure H$_{2}$O volume as a function of time',
                ha='center', va='center', color='white', fontsize=20)

# Save & display the plot
pyplot.savefig('../reports/figures/graphs/pure_H2O_vol_vs_time.png', bbox_inches='tight')
pyplot.show()
```
